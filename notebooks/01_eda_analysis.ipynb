# Data Cleaning and preparation
This section focuses on preparing the data for modelling,by handling missig values, correcting data types and removing unnecesary features

```python
df.isna().mean().sort_values(ascending=False)
```

Insights
- it is Obvious that the deck column is too incomplete to keep as no reasonable insights can be derived from it due to the population of missing values

```python
df_clean = df.drop(columns =["deck"])
```

```python
df_clean["age"] = df_clean["age"].fillna(df_clean["age"].median())
```

```python
df_clean["embarked"] = df_clean["embarked"].fillna(df_clean["embarked"].mode()[0])
```

```python
df_clean["embark_town"] = df_clean["embark_town"].fillna(df_clean["embark_town"].mode()[0])
```

```python
df_clean.isna().mean().sort_values(ascending=False)
```

```python
df_clean.info()
```

```python
df_clean["survived"] = df_clean["survived"].astype("category")
df_clean["sex"] = df_clean["sex"].astype("category")
df_clean["embarked"] = df_clean["embarked"].astype("category")
```

```python
df_clean.info()
```

## Data Preparation Insights
- Columns with excessive missing values were removed to avoid noisy imputations.
- Numerical features were imputed using median values to reduce outlier influence.
- Categorical variables were explicitly cast to appropriate types for downstream modeling.
